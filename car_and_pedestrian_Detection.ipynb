```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

results = model.train(
    data="/content/final6/data.yaml",
    epochs=100,
    imgsz=640,
    batch=8,
    patience=10
)
```

```python
metrics = model.val()
print(metrics)
```

```python
from ultralytics import YOLO
import pandas as pd

model = YOLO('/content/runs/detect/train/weights/best.pt')

metrics = model.val()

# Class-wise values
precision = metrics.box.p
recall = metrics.box.r
map50_class = metrics.box.maps   # this gives class-wise mAP@0.5:0.95

data = {
    "Class": ["Car", "Pedestrian"],
    "Precision": [round(precision[0],3), round(precision[1],3)],
    "Recall": [round(recall[0],3), round(recall[1],3)],
    "mAP@0.5:0.95": [round(map50_class[0],3), round(map50_class[1],3)]
}

df = pd.DataFrame(data)

df
```
